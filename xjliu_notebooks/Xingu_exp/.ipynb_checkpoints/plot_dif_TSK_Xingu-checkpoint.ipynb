### How many ensemble members are enough? 

```python
fig,axes = plt.subplots(3,2,figsize=(14,12),sharex=True,sharey=True)
axes = axes.ravel()

im0 = axes[0].contourf(lons,lats,dif[0:3,:,:].mean('record'),cmap='PiYG',levels=np.arange(-300,301,50))
im1 = axes[1].contourf(lons,lats,dif[3:6,:,:].mean('record'),cmap='PiYG',levels=np.arange(-300,301,50))
im2 = axes[2].contourf(lons,lats,dif[0:6,:,:].mean('record'),cmap='PiYG',levels=np.arange(-300,301,50))
im3 = axes[3].contourf(lons,lats,dif[6:12,:,:].mean('record'),cmap='PiYG',levels=np.arange(-300,301,50))
im2 = axes[4].contourf(lons,lats,dif[0:12,:,:].mean('record'),cmap='PiYG',levels=np.arange(-300,301,50))
im3 = axes[5].contourf(lons,lats,dif[0:19,:,:].mean('record'),cmap='PiYG',levels=np.arange(-300,301,50))
fig.colorbar(im0,ax=axes[0],label='[mm]')
fig.colorbar(im1,ax=axes[1],label='[mm]')
fig.colorbar(im2,ax=axes[2],label='[mm]')
fig.colorbar(im3,ax=axes[3],label='[mm]')
fig.colorbar(im2,ax=axes[4],label='[mm]')
fig.colorbar(im3,ax=axes[5],label='[mm]')

titles=['0-3','3-6','0-6','6-12','0-12','12-19']
for ii in range(6):
    axes[ii].set_title('Members '+titles[ii])
```

### Evolution of precipitation and cloud fraction, using ensemble mean of all 19 members

```python
precip_obs_ensmean = ds0['RAINNC'][:,:,:,:].mean(('record','south_north','west_east'))
precip_notrees_ensmean = ds1['RAINNC'][:,:,:,:].mean(('record','south_north','west_east'))
#del precip_obs
#del precip_notrees
```

```python
dif_ensmean = precip_notrees_ensmean - precip_obs_ensmean
dif_ensmean_daily = np.empty(20)
for ii in range(20):
    dif_ensmean_daily[ii]=dif_ensmean[ii*24] ## still accumulated value
plt.plot(np.arange(10,29),dif_ensmean_daily[1:]-dif_ensmean_daily[0:-1])
plt.title('No trees minus Observed trees')
plt.xlabel('Days since start of the simulation')
plt.ylabel('Daily Accumualted Precip [mm]')
plt.plot([10,28],[0,0],'--k')
```
