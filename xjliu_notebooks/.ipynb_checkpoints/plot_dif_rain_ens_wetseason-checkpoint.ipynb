```python
### OBS land use
path = '/jet/home/xjliu/Tanguro_exp/'
files = sorted(glob.glob(path+'Tanguro201401/3Dfields_d01_2014-01-*00:00:00'))
with xarray.open_dataset(files[0],decode_times=True,
                         mask_and_scale='True') as ds:
    Rainnc_obs = ds['RAINNC']
    Rainsh_obs = ds['RAINSH']

for filename in files[1:]:
    with xarray.open_dataset(filename,decode_times=True,
                         mask_and_scale='True') as ds:
        Rainnc_obs = xarray.concat([Rainnc_obs,ds['RAINNC']],dim='Day')
        Rainsh_obs = xarray.concat([Rainsh_obs,ds['RAINSH']],dim='Day')

Rainnc_obs_sum = Rainnc_obs.sum('Time').mean('Day')
Rainsh_obs_sum = Rainsh_obs.sum('Time').mean('Day')


### No trees
files = sorted(glob.glob(path+'Tanguro201401_notrees/3Dfields_d01_2014-01-*00:00:00'))
with xarray.open_dataset(files[0],decode_times=True,
                         mask_and_scale='True') as ds:
    Rainnc_notrees = ds['RAINNC']
    Rainsh_notrees = ds['RAINSH']

for filename in files[1:]:
    with xarray.open_dataset(filename,decode_times=True,
                         mask_and_scale='True') as ds:
        Rainnc_notrees = xarray.concat([Rainnc_notrees,ds['RAINNC']],dim='Day')
        Rainsh_notrees = xarray.concat([Rainsh_notrees,ds['RAINSH']],dim='Day')
Rainnc_notrees_sum = Rainnc_notrees.sum('Time').mean('Day')
Rainsh_notrees_sum = Rainsh_notrees.sum('Time').mean('Day')


### All trees
files = sorted(glob.glob(path+'Tanguro201401_alltrees/3Dfields_d01_2014-01-*00:00:00'))
with xarray.open_dataset(files[0],decode_times=True,
                         mask_and_scale='True') as ds:
    Rainnc_alltrees = ds['RAINNC']
    Rainsh_alltrees = ds['RAINSH']

for filename in files[1:]:
    with xarray.open_dataset(filename,decode_times=True,
                         mask_and_scale='True') as ds:
        Rainnc_alltrees = xarray.concat([Rainnc_alltrees,ds['RAINNC']],dim='Day')
        Rainsh_alltrees = xarray.concat([Rainsh_alltrees,ds['RAINSH']],dim='Day')
        
Rainnc_alltrees_sum = Rainnc_alltrees.sum('Time').mean('Day')
Rainsh_alltrees_sum = Rainsh_alltrees.sum('Time').mean('Day')
```
