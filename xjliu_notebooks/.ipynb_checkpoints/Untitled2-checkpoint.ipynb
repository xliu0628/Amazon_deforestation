```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import time
import matplotlib.patches as patches
import glob
```

```python
files = glob.glob('/ocean/projects/ees210014p/xjliu/DATA/TRMM/*.nc4')
```

```python
aa = xr.open_dataset(files[0])  
prec = aa['precipitation'].transpose('nlat','nlon')
```

```python
for ii in files[1:]:
    tmp = xr.open_dataset(ii)
    prec = xr.concat([prec,tmp['precipitation'].transpose('nlat','nlon')],dim='time')
```

```python
prec_4D = np.reshape(prec.values,(22,12,260,302))
```

```python
prec_areaave = prec[:,150:230,100:200].mean(('nlat','nlon'))
prec_mean = prec_4D[:,1:3,:,:].mean(0)
```

```python
plt.contourf(prec_4D[:,1,:,:].mean(0),cmap='jet')
plt.grid()
```

```python
plt.plot(prec_areaave)
```

```python
prec_ts = prec_areaave[:22]
for ii in range(21):
    prec_ts[ii] = prec_areaave[ii*12+10:ii*12+14].mean()
```
