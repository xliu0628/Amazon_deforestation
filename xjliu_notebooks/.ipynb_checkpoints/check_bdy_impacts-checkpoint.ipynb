```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import xarray
import glob
import matplotlib.colors as colors
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
import palettable
from palettable.colorbrewer.sequential import YlOrRd_9,RdPu_9
from palettable.colorbrewer.diverging import RdYlGn_10_r
from wrf import getvar, ALL_TIMES
from netCDF4 import Dataset
import subprocess
```

### Rainfall data

```python
path = '/jet/home/xjliu/Tanguro_exp/'

with xarray.open_dataset(path+'Tanguro201401/wrfout_d01_2014-01-01_00:00:00',decode_times=True,
                         mask_and_scale='True') as ds:
    lats = ds['XLAT'][0,:,:]
    lons = ds['XLONG'][0,:,:]

file1 = path+'post_processing/precip/precip_wetseason_obs.nc'
file2 = path+'post_processing/precip/precip_wetseason_notrees.nc'

simuls = ['Tanguro201401','Tanguro201402','Tanguro201501','Tanguro201502','Tanguro201512',
         'Tanguro201601','Tanguro201602','Tanguro201612']

with xarray.open_dataset(file1,decode_times=True,
                         mask_and_scale='True') as ds1:
    print('ok')
    
with xarray.open_dataset(file2,decode_times=True,
                         mask_and_scale='True') as ds2:
    print('ok')
dif = ds2['RAINNC']-ds1['RAINNC']

fig,axes = plt.subplots(4,2,figsize=(9,14),sharex=True,sharey=True)
#fig.subplots_adjust(wspace=0.275)
axes = axes.ravel()
for ii in range(8):
    im = axes[ii].contourf(lons,lats,dif[ii,1,:,:],cmap='PiYG',
                          levels = np.linspace(-400,400,17),extend='both')

    axes[ii].set_title(simuls[ii])
    fig.colorbar(im,ax=axes[ii],label='[mm]')
```

```python
path = '/jet/home/xjliu/Tanguro_exp/'
file1 = path+'post_processing/precip_dry/precip_dry_all_obs.nc'
file2 = path+'post_processing/precip_dry/precip_dry_all_notrees.nc'
simuls = ['Tanguro201406','Tanguro201407','Tanguro201408','Tanguro201506','Tanguro201507','Tanguro201508',
         'Tanguro201606','Tanguro201607','Tanguro201608']

fig,axes = plt.subplots(2,4,figsize=(16,7),sharex=True,sharey=True)
axes = axes.ravel()

for ii in range(8):
    file1 = path+'post_processing/precip_dry/RAINNC_'+simuls[ii]
    file2 = file1+'_notrees'
   # print(file1)
    with xarray.open_dataset(file1,decode_times=True,
                             mask_and_scale='True') as ds1:
        print('ok')
    
    with xarray.open_dataset(file2,decode_times=True,
                             mask_and_scale='True') as ds2:
        print('ok')
    dif = ds2['RAINNC']-ds1['RAINNC']
    im = axes[ii].contourf(lons,lats,dif[-1,:,:],cmap='PiYG',levels=np.arange(-10,11,1),
                          extend='both')
    fig.colorbar(im,ax=axes[ii],label='[mm]')
    axes[ii].set_title(simuls[ii])

```

```python
path = '/jet/home/xjliu/Tanguro_exp/'

with xarray.open_dataset(path+'Tanguro201401/wrfout_d01_2014-01-01_00:00:00',decode_times=True,
                         mask_and_scale='True') as ds:
    lats = ds['XLAT'][0,:,:]
    lons = ds['XLONG'][0,:,:]

file1 = path+'post_processing/precip_obs_28.nc'
file2 = path+'post_processing/precip_notrees_28.nc'

with xarray.open_dataset(file1,decode_times=True,
                         mask_and_scale='True') as ds1:
    print('ok')
    
with xarray.open_dataset(file2,decode_times=True,
                         mask_and_scale='True') as ds2:
    print('ok')
dif = ds1['RAINNC']-ds2['RAINNC']
dif10 = dif[:10,-1,:,:].mean('record')
dif8 = dif[10:,-1,:,:].mean('record')
dif4 = dif[14:,-1,:,:].mean('record')
dif2 = dif[16:,-1,:,:].mean('record')
dif1 = dif[10,-1,:,:]#.mean('record')
```

```python
fig,axes = plt.subplots(2,2,figsize=(10,9),sharex=True,sharey=True)
#fig.subplots_adjust(wspace=0.275)
axes = axes.ravel()
for ii,num in enumerate([10,8,4,2]):
    im0 = axes[0].contourf(lons,lats,dif10,cmap='PiYG')
                          #levels = np.linspace(-400,400,17))
    im1 = axes[1].contourf(lons,lats,dif8,cmap='PiYG')
    im2 = axes[2].contourf(lons,lats,dif4,cmap='PiYG')
    im3 = axes[3].contourf(lons,lats,dif1,cmap='PiYG')


fig.colorbar(im0,ax=axes[0],label='[mm]')
fig.colorbar(im1,ax=axes[1],label='[mm]')
fig.colorbar(im2,ax=axes[2],label='[mm]')
fig.colorbar(im3,ax=axes[3],label='[mm]')
```
