```python
path = '/jet/home/xjliu/Tanguro_exp/'

with xarray.open_dataset(path+'Tanguro201401/wrfout_d01_2014-01-01_00:00:00',decode_times=True,
                         mask_and_scale='True') as ds:
    lats = ds['XLAT'][0,:,:]
    lons = ds['XLONG'][0,:,:]

file1 = path+'post_processing/precip/precip_wetseason_obs.nc'
file2 = path+'post_processing/precip/precip_wetseason_notrees.nc'

simuls = ['Tanguro201401','Tanguro201402','Tanguro201501','Tanguro201502','Tanguro201512',
         'Tanguro201601','Tanguro201602','Tanguro201612']

with xarray.open_dataset(file1,decode_times=True,
                         mask_and_scale='True') as ds1:
    print('ok')
    
with xarray.open_dataset(file2,decode_times=True,
                         mask_and_scale='True') as ds2:
    print('ok')
dif = ds2['RAINNC']-ds1['RAINNC']

fig,axes = plt.subplots(4,2,figsize=(9,14),sharex=True,sharey=True)
#fig.subplots_adjust(wspace=0.275)
axes = axes.ravel()
for ii in range(8):
    im = axes[ii].contourf(lons,lats,dif[ii,1,:,:],cmap='PiYG',
                          levels = np.linspace(-400,400,17),extend='both')

    axes[ii].set_title(simuls[ii])
    fig.colorbar(im,ax=axes[ii],label='[mm]')
```
